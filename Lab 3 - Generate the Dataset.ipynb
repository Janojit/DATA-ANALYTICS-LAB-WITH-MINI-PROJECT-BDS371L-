# Code:

```python
# Importing Libraries 

import pandas as pd
import random
```

### Question 1 <br>
**Take a domain on your own, generate your own dataset having -** <br>
- At least 4 Numeric Values<br>
- At least 2 Categorical Values<br>
- At least 2000 Rows<br>

```python
# Creating a dataframe

df1 = pd.DataFrame(columns = ['Employee ID','Age','Working Hour','Typing Speed','Loan','Gender','Salary'])

ln = []
gd = []

count = int(input("Enter Number Of Entries: "))

for i in range(0,count):
    newRow = {}
    newRow['Employee ID'] = random.randint(100000,999999)
    newRow['Age'] = random.randint(20,60)
    newRow['Working Hour'] = random.randint(6,10)
    newRow['Typing Speed'] = random.randint(30,120)
    a = random.uniform(0,1)
    if a<0.3:
        ln.append("Yes")
    else:
        ln.append("No")
    newRow['Loan'] = ln[i]
    if a<0.5:
        gd.append("M")
    elif a>=0.5 and a<0.9:
        gd.append("F")
    else:
        gd.append("O")
    newRow['Gender'] = gd[i]
    newRow['Salary'] = random.randint(50000,1000000)
    df1 = df1.append(newRow,ignore_index=True)

df1['Age'] = df1['Age'].astype('int64')
df1['Working Hour'] = df1['Working Hour'].astype('int64')
df1['Typing Speed'] = df1['Typing Speed'].astype('int64')
df1['Salary'] = df1['Salary'].astype('int64')

df1
```

```python
df1.info()
```

```python
df1.describe()
```

### Obsevation:
**A Employee details based dataframe is generated with 2000 rows, which has 4 numeric and 2 categorical attribute, it is defined as df1. Dictionary is used to store each entries and appended it to the dataframe. Using condition statement randomization has been given weightage.**

### Question 2 A<br>
**Sample 20% of the Dataset and Create a new dataframe**<br>

```python
# Creating a new dataframe using Sampling method

df2 = df1.sample(frac=0.2, replace=True)
df2
```

```python
df2.info()
```

```python
df2.describe()
```

### Observation:
**20% sample dataframe from the main dataframe is generated, which is assigned as df2. Here sample method is used.**

### Question 2 B<br>
**In the sampled dataset, Remove some rows based on some column values**

```python
# Removing some rows based on Column based conditions

df2 = df2.drop(df2[df2['Working Hour']>8].index)
df2
```

### Observation:
**Entries whose working hour is less than 8 hour is deleted. Here drop method is used.**

### Question 2 C<br>
**Replace some values of one column with "NaN" or "Null**

```python
df2.rename({'Typing Speed': 'Typing_Speed'}, axis=1, inplace=True)

df2.Typing_Speed.iloc[df2['Typing_Speed']<50] = None
df2.Gender.iloc[df2['Gender']=="O"] = None
df2.Age.iloc[df2['Age']<30] = None


df2
```

```python
df2.info()
```

```python
df2.isnull().sum()
```

### Observation:
**Based on certain conditions some entries of Typing Speed, Age and Gender is converted to Nan/Null values. Here iloc method is used to fetch those entries.**

### Question 2 D<br>
**Find the difference between .fillna(), .interpolate() and .dropna()**

```python
# Using fillna removing Nan values

df2['Typing_Speed'].fillna(40,inplace=True)
df2
```

```python
df2.isnull().sum()
```

```python
# Using interpolate removing Nan values

df2['Age'] = df2['Age'].interpolate(method='linear',limit_direction = 'both')
df2
```

```python
df2.isnull().sum()
```

```python
# Using dropna removing Nan values

df2 = df2.dropna()
df2
```

```python
df2.isnull().sum()
```

### Observation:
- Difference between .fillna(), .interpolate() and .dropna(),
 - .fillna(): This method is used to replace Null values with a specific value. Here fillna method is used to replace Null values of Typing Speed column to 40.
 - .interpolate(): This method is used to replace Null values with numerics between previous and next row. Here interpolate method is used to replace Null values of Age Attribute.
 - .dropna(): This method is used to remove all rows/entries who have any Null value.
