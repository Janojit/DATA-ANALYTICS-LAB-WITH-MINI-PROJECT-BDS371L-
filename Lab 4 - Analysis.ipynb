# Code:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("export_dataframe.csv")
df
```

```python
df.describe()
```

```python
df.info()
```

```python
sns.set(style='whitegrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [8, 5]})
sns.distplot(
    df['Age'], norm_hist=False, kde=False, bins=20, hist_kws={"alpha": 1}
).set(xlabel='Age', ylabel='Count');
```

### Obsevation:
From displot of Age, we observed that most of the employees age is 60. So the organization needs to hire more freshers as most of the employees going to retire.

```python
sns.set(style='whitegrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [8, 5]})
sns.distplot(
    df['Working Hour'], norm_hist=False, kde=False, bins=20, hist_kws={"alpha": 1}
).set(xlabel='Working Hour', ylabel='Count');
```

### Observation:
From the displot of Working Hour, we observed that most of employees are working for 5 to 6 hours, which may caused low productivity as a whole organization that's need to focused.

```python
sns.set(style='whitegrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [8, 5]})
sns.distplot(
    df['Typing Speed'], norm_hist=False, kde=False, bins=20, hist_kws={"alpha": 1}
).set(xlabel='Typing Speed', ylabel='Count');
```

### Observation: 
From the displot of Typing speed, we can conclude that the organization doesn't focuses much on Typing Speed while hiring employees.

```python
sns.countplot(df['Gender']);
```

### Observation:
From the count plot of Gender, we observed that most of the employees of this organization are Male. So, maybe the previous hiring process was biased against Females and Others.

```python
sns.countplot(df['Loan']);
```

### Observation:
From the count plot of Loan, we observed that most the employees are without any loan. It is a good sign for the organization as it indicates most of the employees are getting their actual wage.

```python
ax = sns.catplot(x="Gender", y="Age", kind="violin",
                 split=True, palette="pastel", data=df, height=4.2, aspect=1.4)
```

### Observation:
From the catplot of Gender and Age, we observed that median age of females are higher as compare to males and others. Now we previously seen that number of female is also less and a huge number of employees are going to retire. So there is a huge probability that a large set of female employees are going to retire. From which we can suggest that the organization should recruit more numbers of female freshers.

```python
sns.catplot(x="Gender", y="Salary", kind="box", data=df, height=4.2, aspect=1.4)
```

### Observation:
From boxplot of Gender and Salary, we observed that most of the males are getting more salaries as compare to females and others. We have previously seen that already the hiring process of this organization is biased against females and others now we also observed the biasness in terms of salary also. Organization needs to focus on this issue as it depicts negativity towards the organization.

```python
sns.scatterplot(x=df['Typing Speed'], y=df['Salary']);
```

```python
sns.jointplot(x=df['Working Hour'], y=df['Salary']);
```

```python
sns.barplot(df["Working Hour"],df["Salary"])
```

### Observation:
From the barplot of Working Hour and Salary, we observed those who are working less hours are getting more salaries, which is unusual. Maybe less hours employees are more efficient thats why they needs less time to complete a task and that's the reason they are getting more salaries.

```python
sns.barplot(df["Loan"],df["Salary"])
```

### Observation:
From the bar plot of Loan and salary, we observed that employees who are under loan are getting enough salary. It is a good sign for the organization as it indicates most of the employees are getting their actual wage.

```python
sns.stripplot(x ='Loan', y ='Gender', data = df,
              jitter = True, dodge = True)
```

### Observation:
From the strip plot of Loan and salary, we observed that employees who are under loan are male. As we also observed males are getting higher salary maybe that is the reason they are taking more loans.

```python
g = sns.PairGrid(df, y_vars=["Loan"], x_vars=["Age", "Salary"], height=4.5, aspect=1.1)
g.map(plt.scatter, alpha=0.6)
```

```python
sns.pairplot(data = df, vars=['Loan','Gender','Salary','Age','Working Hour','Typing Speed'])
plt.show()
```

```python
df1 = df[['Age','Working Hour','Typing Speed','Salary']].corr()

df1.corr()

sns.heatmap(df1.corr(), annot=True, cmap = 'Reds')
plt.show()
```

### Observation:
Heatmap shows the correlation between all numeric attributes.
