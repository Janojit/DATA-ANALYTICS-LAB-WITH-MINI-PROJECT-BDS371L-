```python
from prettytable import PrettyTable


class Stock:
    
    def __init__(self,stockName,observationTime,openValue,highValue,lowValue,currentValue):
        self.stockName = stockName
        self.observationTime = observationTime
        self.openValue = openValue
        self.highValue = highValue
        self.lowValue = lowValue
        self.currentValue = currentValue
        
    def setStockName(self,name):
        self.stockName = name
    
    def setOpenValue(self,openV):
        self.openValue = openV
        
    def setHighValue(self,highV):
        self.highValue = highV
        
    def setLowValue(self,lowV):
        self.lowValue = lowV
        
    def setCurrentValue(self,currentV):
        self.currentValue = currentV
        
    def getStockName(self):
        return self.stockName
        
    def getObservationTime(self):
        return self.observationTime
    
    def getOpenValue(self):
        return self.openValue
    
    def getHighValue(self):
        return self.highValue
    
    def getLowValue(self):
        return self.lowValue
    
    def getCurrentValue(self):
        return self.currentValue
    
    def barChart(self):
        val = [self.openValue,self.highValue,self.lowValue,self.currentValue]
        l1 = ["Open   ","High   ","Low    ","Current"]
        v = 0
        for i in val:
            print("\n",end=" ")
            print(l1[v],end='|')
            v+=1
            for j in range(round(i/1000)):
                print("#",end=" ")
            print(" ",i,"$",end=" ")
        
    def display(self):
        global stockPrice
        
        print("\t\t\tstockPrice")
        
        stocksPrice = [self.getStockName(),self.getObservationTime(),self.getOpenValue(),self.getHighValue(),self.getLowValue(),self.getCurrentValue()]
        
        stockPrice = PrettyTable()
        
        stockPrice.field_names = ["Name","Observation Time","Open Value","High Value","Low Value","Current Value"]
        
        stockPrice.add_row(stocksPrice)
        
        print(stockPrice)
```

```python
def displayAll(listName):
    print("\t\t\tProduct Details")
    
    stockPriceAll = PrettyTable()
    
    stockPriceAll.field_names = ["Name","Observation Time","Open Value","High Value","Low Value","Current Value"]
    
    for x in listName:
        stoPrice= [x.getStockName(),x.getObservationTime(),x.getOpenValue(),x.getHighValue(),x.getLowValue(),x.getCurrentValue()]
        stockPriceAll.add_row(stoPrice)
        
    print(stockPriceAll)
```

```python
def checkHigh():
    for x in allStock:
        if x.currentValue>x.highValue:
            highV = x.currentValue
            x.setHighValue(highV)
            print('\033[1m'+"Checking Done!!!"+'\033[0m')
            print('\033[1m'+"High Value Changed!!!"+'\033[0m')
            x.display()
```

```python
def checkLow():
    for x in allStock:
        if x.lowValue>x.openValue:
            lowV = x.openValue
            x.setLowValue(lowV)
            print('\033[1m'+"Checking Done!!!"+'\033[0m')
            print('\033[1m'+"Low Value Changed!!!"+'\033[0m')
            x.display()
```

```python
allStock = []
```

```python
def updateDetails():
    stockName = input("Enter Stock Name Which Details Will Be Updated: ")
    for x in allStock:
        if x.stockName==stockName:
            name = input("Update Stock Name: ")
            openV = int(input("Update Open Value: "))
            highV = int(input("Update High Value: "))
            lowV = int(input("Update Low Value: "))
            currentV = int(input("Update Current Value: "))
            x.setStockName(name)
            x.setOpenValue(openV)
            x.setHighValue(highV)
            x.setLowValue(lowV)
            x.setCurrentValue(currentV)
            x.display()
```

```python
def displayHighestValue():
    hig = 0
    for x in allStock:
        if x.openValue>hig:
            hig = x.openValue
    
    openValue = hig
    for x in allStock:
        if x.openValue==openValue:
            x.display()
```

```python
def displayOldestValue():
    l1 = []
    for x in allStock:
        time = x.observationTime
        l1.append(time)
    l1.sort()
    old = l1[0]
    for x in allStock:
        if x.observationTime == old:
            x.display()
```

```python
def toStrToDic():
    for x in allStock:
        StockDic = {}
        StockDic['Current Value'] = x.getCurrentValue()
        print(StockDic)
```
