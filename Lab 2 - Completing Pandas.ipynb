```python
# Importing Libraries 

import pandas as pd
import numpy as np
import matplotlib as plt
```

### Question 1

```python
# Importing Titanic Dataset

df = pd.read_csv("titanic.csv")

df
```

### Observation: 
**To import a csv file we need to use "read_csv" method which will automatically generate Dataframe from that csv file.**

### Question 2

```python
# Renaming Attributes

df.rename({'Survived': 'Survival_Status'}, axis=1, inplace=True)
df.rename({'Pclass': 'Class'}, axis=1, inplace=True)
df.rename({'Siblings/Spouses Aboard': 'Siblings/Spouses_Aboard'}, axis=1, inplace=True)
df.rename({'Parents/Children Aboard': 'Parents/Children_Aboard'}, axis=1, inplace=True)
df
```

### Obsevation:
**To rename an attribute we need "rename" method and apart from that we need to also define the axis.**

### Question 3

```python
# Converting fare from USD to INR

df['Fare'] = df['Fare']*79.19
df
```

### Exploration of Survival Status

```python
print(25*"-","\033[1m","Exploration Of Survival Status","\033[0m",25*"-")
print("")
print("Average Age of Survived Passengers: ",df[df["Survival_Status"]==1]["Age"].mean())
print("Average Age of Died Passengers: ",df[df["Survival_Status"]==0]["Age"].mean())
print("Average Fare of Survived Passengers: ",df[df["Survival_Status"]==1]["Fare"].mean())
print("Average Fare of Died Passengers: ",df[df["Survival_Status"]==0]["Fare"].mean())
print("Average Class of Survived Passengers: ",df[df["Survival_Status"]==1]["Class"].mean())
print("Average Class of Died Passengers: ",df[df["Survival_Status"]==0]["Class"].mean())
print("Male Survival Rate: ",(df[df["Sex"]=="male"]["Survival_Status"].mean())*100,"%")
print("Female Survival Rate: ",(df[df["Sex"]=="female"]["Survival_Status"].mean())*100,"%")
print("Survival Rate of Whose Relatives Lives In Abroad: ",df[(df["Siblings/Spouses_Aboard"]>0) | (df["Parents/Children_Aboard"]>0)]["Survival_Status"].mean())
print("Survival Rate of Whose Relatives Doesn't Live In Abroad: ",df[(df["Siblings/Spouses_Aboard"]==0) | (df["Parents/Children_Aboard"]==0)]["Survival_Status"].mean())
```

### Observation:
- **As per the survival status we can infer that average age of Survival Passengers are almost same as compare to died passangers which is quite normal.**
- **We observe that people who purchased more fare tickets are survived more as compare to others, which is indicating that maybe more fare paid passengers are richer thats why maybe they bribed during rescue process or the rescue process was biased and prioritize based on wealth of the passengers.**
- **Female survival rate very high as compare to male survival rate, from which we conclude that the rescue process was prioritize female.**

### Fare Exploration

```python
print(25*"-","\033[1m","Fare Exploration","\033[0m",25*"-")
print("")
print("Average Fare of All Passengers: ",df["Fare"].mean())
print("Average Fare of Less Than 18 Aged Passengers: ",df[df["Age"]<18]["Fare"].mean())
print("Average Fare of More Than 18 Aged Passengers: ",df[df["Age"]>18]["Fare"].mean())
print("Average Fare of Male Passengers: ",df[df["Sex"]=="male"]["Fare"].mean())
print("Average Fare of Female Passengers: ",df[df["Sex"]=="female"]["Fare"].mean())
print("Average Fare of Mr. Whose Relatives Lives In Abroad: ",df[(df["Name"].str.find("Mr.")==0) & ((df["Siblings/Spouses_Aboard"]>0) | (df["Parents/Children_Aboard"]>0))]["Fare"].mean())
print("Average Fare of Mr. Whose Relatives Does Not Live In Abroad: ",df[(df["Name"].str.find("Mr.")==0) & ((df["Siblings/Spouses_Aboard"]==0) | (df["Parents/Children_Aboard"]==0))]["Fare"].mean())
print("Average Fare of Mrs. Whose Relatives Lives In Abroad: ",df[(df["Name"].str.find("Mrs.")==0) & ((df["Siblings/Spouses_Aboard"]>0) | (df["Parents/Children_Aboard"]>0))]["Fare"].mean())
print("Average Fare of Mrs. Whose Relatives Does Not Live In Abroad: ",df[(df["Name"].str.find("Mrs.")==0) & ((df["Siblings/Spouses_Aboard"]==0) | (df["Parents/Children_Aboard"]==0))]["Fare"].mean())
```

### Observation:
- **Average female fare is much higher than male, which indicates that maybe there is large section of male, who purchased low priced tickets or they are travelling without fare.**
- **People who have relative in abroad are much richer thats why they booked high priced tickets.**
- **Average fare price of Adults and non-adults are more or less similar.**

### Question 5 (CRUD Opperations)

### Row Wise

```python
# Creating a new row

df1 = df
df1 =  df1.append({'Survival_Status':0, 'Class':3, 'Name':'Janojit Chakraborty','Sex':'male','Age':18.0,'Siblings/Spouses_Aboard':0,'Parents/Children_Aboard':0,'Fare': 3500.000}, ignore_index = True)
df1
```

```python
# reading a row using the index

df1.loc[89]
```

```python
# Updating the row

df1.at[887,'Survival_Status']=1
df1
```

```python
# deleting a row from the dataframe

df1 = df1.drop(887)
df1
```

### Observation:
- **I used append to add new row in the dataframe and it always by default added at the end.**
- **loc is used for row indexing and at for row updation and drop for row deletion.**
