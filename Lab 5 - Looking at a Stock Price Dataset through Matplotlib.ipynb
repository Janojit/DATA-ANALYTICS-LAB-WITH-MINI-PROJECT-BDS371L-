### Question: Find the correlation between the features. Make a Correlation Heatmap. Write your observations about the same.

```python
plt.figure(figsize=(25,15))
sns.heatmap(stockprice_one_year.corr(method='pearson'), annot=True)
```

### Observation:
As per the heatmap we can conclude that Prev Close, Open, High, Low, Last, Close, VWAP this are highly correlated to each other and Volume, Turnover, Trades, Deliverable Volume, %Deliverble are also highly correlated to each other.

### Question: How can you plot multiple features in a single plot? For instance, use a red line for "LOW", green line for "HIGH", and black line for "CLOSE"

```python
stockprice_one_year.plot(x="Date", y=["Low","High","Close","Open","Last"],color = ['r','g','k','b','y'], figsize = [25,15])
```

### Observation:
To provide different colours we need to use color parameter within the .plot() method.

### Question: What does plt.fill_between() do? can you shade the region between LOW and HIGH?

```python
stockprice_one_year.plot(x="Date", y=["Low","High"],color = ['r','b'], figsize = [25,15])
plt.show()
```

```python
stockprice_one_year.plot(x="Date", y=["Low","High"],color = ['r','b'], figsize = [25,15])
plt.fill_between(stockprice_one_year['Date'], stockprice_one_year['Low'], stockprice_one_year['High'], color='k',alpha=0.5)
plt.show()
```

### Observation:
Using fill_between we can fill any colour in between two plotted lines.

### Question: Take the CLOSE price, can you give a colour for the first half of the year, and a different colour for second half?

```python
x = stockprice_one_year.loc[0:125]
y = stockprice_one_year.loc[125:249]

x['Close'].plot(color='red')
y['Close'].plot(color='blue')
```

### Observation:
Using .loc I split the Dataset and using .plot() I plotted it with different colour.
