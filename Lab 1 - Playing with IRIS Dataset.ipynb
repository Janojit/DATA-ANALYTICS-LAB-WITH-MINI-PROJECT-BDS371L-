# Code:

```python
# Importing Libraries 

import pandas as pd
import numpy as np
import matplotlib as plt
```

### Question 1

```python
# Importing IRIS Dataset

from sklearn import datasets

iris = datasets.load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])

df["target"] = iris.target

target_names = {0: "Iris-Setosa", 1: "Iris-Versicolour", 2:"Iris-Virginica" }

df['target'] =df['target'].map(target_names)

df
```

**Observation** : First I converted the dataset into dataFrame using Pandas. Then I have mapped target values 0 as Iris-Setosa, 1 as Iris-Versicolour and 2 as Iris-Virginica.

### Question 2

```python
# Printing the Head of the Dataset

df.head()
```

```python
# Printing the Tail of the Dataset

df.tail()
```

**Observation** : I found head and tail which means the first and last 5 observation using Pandas.

### Question 3

```python
# Creating a New Column named "Average" and find the Average of existing four features and store the same

df['Average'] = (df['sepal length (cm)'] + df['sepal width (cm)'] + df['petal length (cm)'] + df['petal width (cm)'])/4

df
```

**Observation**: I created an Attribute "Average" and input values based on the average of all four features in each entries.<br>

### Question 4

```python
# Creating a New Column named "Sum" and find the Sum of existing four features and store the same

df['Sum'] = (df['sepal length (cm)'] + df['sepal width (cm)'] + df['petal length (cm)'] + df['petal width (cm)'])

df
```

**Observation** : I created an Attribute "Sum" and input values based on the summation of all four features in each entries.

### Question 5

```python
# Checking how many entires of "IRIS Setosa", "IRIS Virgnica" and "IRIS Versicolor" are present in the Dataset

df.target.value_counts()
```

**Observation** : I used a pandas command "value_counts()" and found how many entries as per each categorical variable.

### Question 6

```python
# Finding the Descriptive Statistics about all the Six Columns (existing four + two made-up features)

df.describe()
```

### Question 7

```python
# Finding the entries whose Sum Values are more than a User Input

n = int(input("Enter A INTEGER Value: "))

df[df['Sum']>n]
```

### Question 8

```python
# Findinng the entries whose Sepal_Length is greater than Petal_Width


df[df['sepal length (cm)']>df['petal width (cm)']]
```
