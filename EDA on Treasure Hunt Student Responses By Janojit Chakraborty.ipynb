## Data Visualization

```python
sns.countplot(x=df['Class'])
plt.xticks(rotation=90)
```

**Observation:**<br>
As per the Countplot, we observe that the Students of 3BSc EA have shown more active participation in the Event as compare to 3 BSc DS.

```python
sns.countplot(x=df['Gender'])
plt.xticks(rotation=90)
```

**Observation:**<br>
Both Genders have equally participated in the Event.

```python
sns.countplot(x=df.iloc[:,5]).set(xlabel='Communication Type', ylabel='Count');
plt.xticks(rotation=90)
```

**Observation:**<br>
In the event, most of students Verbally communicated with their team mates. Somes of them also used Whatsapp and Phone Calls to communicate with their team mates.

```python
sns.countplot(x=df.iloc[:,6]).set(xlabel='Final Status', ylabel='Count');
plt.xticks(rotation=90)
```

**Observation:**<br>
Most of the students are "Out after Prelims" and less number of students reached in "Final Round", which is natural. So we can conclude that the Event was properly organized by the organizers.

```python
sns.set(style='whitegrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [8, 5]})
sns.distplot(
    df.iloc[:,7], norm_hist=False, kde=False, bins=20, hist_kws={"alpha": 1}
).set(xlabel='Event Rating', ylabel='Count');
```

**Observation:**<br>
The event overall got Rating 4 from most of the participants apart from this some has given Rating 5 and Rating 3 also. From which we can conclude that the Event was successfull as per Ratings.

```python
sns.set(style='whitegrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [8, 5]})
sns.distplot(
    df.iloc[:,8], norm_hist=False, kde=False, bins=20, hist_kws={"alpha": 1}
).set(xlabel='Difficulty Level', ylabel='Count');
```

**Observation:**<br>

```python
sns.barplot(df.iloc[:,6],df.iloc[:,13]).set(xlabel='Final Place', ylabel='Same Language Spoken');
plt.xticks(rotation=90)
```

```python
teamMatesClass = pd.DataFrame(columns = ['Class','Student Count'])

stuCount = []
stuClass = ['1 BSc EA','1 BSc DS','3 BSc EA','3 BSc DS','5 BSc EA','5 BSc DS','MSc']

for i in range(14,21):
    stuCount.append(df.iloc[:,i].sum())
for i in range(0,7):
    teamMateClass = {}
    teamMateClass['Class'] = stuClass[i]
    teamMateClass['Student Count'] = stuCount[i]
    teamMatesClass = teamMatesClass.append(teamMateClass,ignore_index=True)
    
teamMatesClass['Student Count'] = teamMatesClass['Student Count'].astype('int64')

sns.barplot(teamMatesClass.iloc[:,0],teamMatesClass.iloc[:,1]).set(xlabel='Class', ylabel='Student Count');
plt.xticks(rotation=90)
```

**Observation:**<br>
As per the Bar Diagram between Student Count and Class, MSc students count is higher in the event because they are not segerated by Year and Course. Amoung BSc Students 3 BSc EA students have shown maximum participation in the Event.
